```python
import pandas as pd
df = pd.read_csv('test.csv').drop(columns ='Unnamed: 0')
```

```python
df = df.append(df.loc[[0,0,0]]).reset_index(drop = True)
df
```

```python
# 당신이 담배를 피우지 않았다면
if df.at[0, 'Smoking'] == 1:
    df.at[1, 'Smoking'] = 0
    # 시각화

# 당신이 과음을 하지 않는다면
if df.at[0, 'AlcoholDrinking'] == 1:
    df.at[2, 'AlcoholDrinking'] = 0

# 잠이 부족한 당신이 수면을 취한다면
if df.at[0, 'SleepTime'] <= 6:
    df.at[3, 'SleepTime'] = 8

# 당신이 정상 범주의 BMI 지수를 가질 경우
if (df.at[0, 'BMI'] < 18.5) or (df.at[0, 'BMI'] > 25):
    df.at[4, 'BMI'] = 22
```
